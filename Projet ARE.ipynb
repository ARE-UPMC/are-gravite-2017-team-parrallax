```python
import numpy as np
from pylab import *
import matplotlib.animation as ani
import matplotlib.pyplot as plt


phi=np.arange(0,2*np.pi,0.01) #arange fonctionne comme range mais avec un pas non entier
#print("Tableau de valeur :", phis) #phis est donc un tableau de valeurs allant de 0 Ã  6.28 espacÃ©es de 0.01
r1 =50 #rayon de l'orbite de 1
tabx1 = r1*np.cos(phi) #tableau des x de l'orbite
taby1 = r1*np.sin(phi) #tableau des y de l'orbite
v1 = 0#vitesse de l'astre 1
time = 0

#"figure" princiale (a peu prÃ¨s la fenetre)
fig = plt.figure(figsize = (8,8), dpi = 100.0, facecolor = "white")



#definit la valeur maximale des axes 

xlim(-100, 100)
ylim(-100, 100)




#gestion des axes

ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))



#tracÃ© de l'orbite de 1
plot(tabx1, taby1)
#tracÃ© de l'astre 1
ax.add_artist(astre1)



#fonctions utilisÃ©es pour l'animation

def init():
    #definition de l'astre1 remplacÃ© par un cercle bleu
    astre1 = plt.Circle((tabx1[0], taby1[0]), 2, color = 'blue', fill = True)
    #tracÃ© de l'astre 1
    ax.add_artist(astre1)
    return astre1
    
def update(i):
    time += 1
    astre1.set_xdata(tabx1[time])
    astre1.set_ydata(taby1[time])
    return astre1
    
#animation

animate = ani.FuncAnimation(fig, update, np.arange(0, len(tabx1)), init_func = init, interval = 20, blit = True)
plt.show()
```
