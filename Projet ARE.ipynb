```python
import numpy as np
from pylab import *
from matplotlib import animation as ani
from matplotlib import pyplot as plt
%matplotlib inline

phi=np.arange(0,2*np.pi,0.01) #arange fonctionne comme range mais avec un pas non entier
#print("Tableau de valeur :", phis) #phis est donc un tableau de valeurs allant de 0 à 6.28 espacées de 0.01
r =50
tabx = r*np.cos(phi)
taby = r*np.sin(phi)
fig = plt.figure(figsize = (8,8), dpi = 72.0, facecolor = "white")
plot(tabx, taby)

#fonctions utilisées pour l'animation

def init():
    alpha = 0
    

def update(*args):
    alpha += 0.01

    
    
#definit la valeur maximale des axes 

xlim(-100, 100)
ylim(-100, 100)


#gestion des axes

ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

#animation

animate = ani.FuncAnimation(fig, func = update, interval = 20, init_func = init)
plt.show()
```
